```python
from tensorflow.keras.utils import img_to_array, load_img
```

```python
img = load_img('/content/elephant.jpg')
```

```python
display(img)
```

```python
x = img_to_array(img)
```

```python
x
```

```python
x.shape
```

```python
x = x.reshape((1,) + x.shape)
```

```python
x.shape
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
```

```python
i = 0
batch_size = 1
for batch in datagen.flow(x, batch_size=batch_size, save_to_dir='/content/augmented_images', save_prefix='elephant', save_format='jpeg'):
    i += 1
    if i > 20:
        break
```
